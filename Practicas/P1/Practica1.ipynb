Librerías utilizadas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
articulos_aceptados = pd.read_csv('Articulos_aceptados.csv')
articulos_enviados = pd.read_csv('Articulos_enviados.csv')
articulos_publicados = pd.read_csv('Articulos_publicados.csv')
cursos_lic = pd.read_csv('Cursos_lic_frente_grupo.csv')
cursos_posgrado = pd.read_csv('Cursos_posgrado_frente_grupo.csv')
cursos_seminario = pd.read_csv('Seminarios_tutorias_posgrado.csv')
tesis_doc_con = pd.read_csv('tesis_doc_con.csv')
tesis_master_con = pd.read_csv('tesis_master_con.csv')
tesis_lic_con = pd.read_csv('tesis_lic_con.csv')
tesis_doc_pro = pd.read_csv('tesis_doc_process.csv')
tesis_master_pro = pd.read_csv('tesis_master_process.csv')
tesis_lic_pro = pd.read_csv('tesis_lic_process.csv')
```

## Articles

```python
anios = [2020, 2021, 2022]
```

```python
suma_2020_acept = np.sum(articulos_aceptados['2020'])
suma_2021_acept = np.sum(articulos_aceptados['2021'])
suma_2022_acept = np.sum(articulos_aceptados['2022'])
suma_T_acept = [suma_2020_acept, suma_2021_acept, suma_2022_acept]
```

```python
suma_2020_env = np.sum(articulos_enviados['2020'])
suma_2021_env = np.sum(articulos_enviados['2021'])
suma_2022_env = np.sum(articulos_enviados['2022'])
suma_T_env = [suma_2020_env, suma_2021_env, suma_2022_env]
```

```python
suma_2020_post = np.sum(articulos_publicados['2020'])
suma_2021_post = np.sum(articulos_publicados['2021'])
suma_2022_post = np.sum(articulos_publicados['2022'])
suma_T_post = [suma_2020_post, suma_2021_post, suma_2022_post]
```

Creamos los histogramas correspondientes a cada tabla

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
fig.suptitle('Academic\'s Articles')

# Accepted
sns.barplot(ax=axes[0], x=anios, y=suma_T_acept)
axes[0].set_title('Accepted Articles')

# Sent
sns.barplot(ax=axes[1], x=anios, y=suma_T_env)
axes[1].set_title('Sent Articles')

# Posted
sns.barplot(ax=axes[2], x=anios, y=suma_T_post)
axes[2].set_title('Posted Articles')
```

## Courses

```python
suma_2020_cursos_lic = np.sum(cursos_lic['2020'])
suma_2021_cursos_lic = np.sum(cursos_lic['2021'])
suma_2022_cursos_lic = np.sum(cursos_lic['2022'])
suma_T_cursos_lic = [suma_2020_cursos_lic, suma_2021_cursos_lic, suma_2022_cursos_lic]
```

```python
suma_2020_cursos_pos = np.sum(cursos_posgrado['2020'])
suma_2021_cursos_pos = np.sum(cursos_posgrado['2021'])
suma_2022_cursos_pos = np.sum(cursos_posgrado['2022'])
suma_T_cursos_pos = [suma_2020_cursos_pos, suma_2021_cursos_pos, suma_2022_cursos_pos]
```

```python
suma_2020_cursos_sem = np.sum(cursos_seminario['2020'])
suma_2021_cursos_sem = np.sum(cursos_seminario['2021'])
suma_2022_cursos_sem = np.sum(cursos_seminario['2022'])
suma_T_cursos_sem = [suma_2020_cursos_sem, suma_2021_cursos_sem, suma_2022_cursos_sem]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
fig.suptitle('Courses given by Academics')

# Accepted
sns.barplot(ax=axes[0], x=anios, y=suma_T_cursos_lic)
axes[0].set_title('Bachelor Courses')

# Sent
sns.barplot(ax=axes[1], x=anios, y=suma_T_cursos_pos)
axes[1].set_title('Master Courses')

# Posted
sns.barplot(ax=axes[2], x=anios, y=suma_T_cursos_sem)
axes[2].set_title('Seminary Courses')
```
