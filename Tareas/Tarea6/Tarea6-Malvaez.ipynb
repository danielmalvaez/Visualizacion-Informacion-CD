```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='darkgrid', palette='pastel', context='notebook', font_scale=1.3)
```

```python
petrol = pd.read_excel('GDP_PETR.xls', sheet_name='Data',skiprows=3)
petrol.head()
```

```python
# Renaming columns
petrol.rename(columns={'Country Name':'country_name', 
                       'Country Code': 'country_id', 
                       'Indicator Name':'indicator_name', 
                       'Indicator Cotde':'indicator_code'}, inplace=True)
```

```python
petrol_new = petrol.dropna(axis=1, how='all')
petrol_new.head()
```

```python
petrol_new.fillna(0, inplace=True)
petrol_new.head()
```

```python
years = [str(x) for x in range(1970, 2020)]
vals = petrol_new.loc[:,'1970':'2020']
vals['petrol_rent_mean'] = vals.mean(axis=1)
values = vals.sort_values(by='petrol_rent_mean', ascending=False)
values.head()
```

```python
first_5 = values.head(5).drop(columns=['petrol_rent_mean'])
first_5 = first_5.transpose()
first_5.columns = [petrol.loc[x]['country_name'] for x in first_5.columns]

first_5.plot(kind='line', figsize=(14,8))
 
plt.title('Tendencia de inmigración de los 5 países principales')
plt.ylabel('Índice petroleo')
plt.xlabel('Años')
plt.show()
```

```python
# My palette colors obtained from https://www.learnui.design/tools/data-color-picker.html
colores = ['#008ccc','#9883df','#f76dba','#ff746e','#ffa600']
sns.color_palette(colores)
```

```python
sns.set_palette(colores)
first_5.plot(kind='area', figsize=(14,8), stacked=True)
 
plt.title('Tendencia de inmigración de los 5 países principales')
plt.ylabel('Índice petroleo')
plt.xlabel('Años')
plt.show()
```

```python
first_5_percentage = first_5.apply(lambda x: x/x.sum(), axis=1)
first_5_percentage.head()
```

```python
years = []
for i in range(1970, 2020, 6):
    years.append(i)
years.append(2020)
```

```python
p = []
for i in range(len(years)-1):
    p.append(Mexico.loc[:,str(years[i]):str(years[i+1])])
```

```python
data = pd.DataFrame()
for i in range(len(p)):
    data[str(years[i])+'-'+str(years[i+1])] = p[i].sum(axis=1)
data
```

```python
plt.figure(figsize=(14,8))
sns.barplot(data.values[0], data.columns)

plt.title('Comportamiento de la renta petrolera mexicana por periodos')
```

```python
d = data.transpose()
d.sort_values(by=154, ascending=False, inplace=True)
periodos_5 = d.head(5).transpose()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0, 0, 0, 0, 0] # relación para cada continente con la que compensar cada cuña.
```

```python
colores = ['#092c42', '#0f3b75','#534087','#8a408c','#bb3f84','#e24672','#fc5d58','#ff7f38','#ffa600']
sns.set_palette(colores)

plt.figure(figsize=(15, 7))
plt.pie(periodos_5.values[0], labels=None,
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            pctdistance=1.12,    # la relación entre el centro de cada sector circular y el inicio del texto generado por autopct
                            colors=colors_list,  # añadir colores personalizados
                            explode=explode_list # "explotar" los 3 continentes más bajos
                            )
plt.axis('equal') 
plt.title('Inmigrantes por continentes en el año 2013', y = 1.12)
plt.legend(labels = data.columns, loc = 'upper left')
plt.show()
```
